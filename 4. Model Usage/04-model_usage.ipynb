The answer would be `0.00105`.

````python
from PIL import Image
import requests
from io import BytesIO
from transformers import ViTFeatureExtractor, ViTForImageClassification

# Step 1: Accept Image URL Input
image_url = "https://images.unsplash.com/photo-1596392927852-2a18c336fb78?ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D&auto=format&fit=crop&w=1280&q=80"

# Step 2: Display the Image
response = requests.get(image_url)
img = Image.open(BytesIO(response.content))
img.show()

# Step 3: Model Loading and Prediction
# Load the pre-trained model and feature extractor
model = ViTForImageClassification.from_pretrained('nateraw/vit-age-classifier')
transforms = ViTFeatureExtractor.from_pretrained('nateraw/vit-age-classifier')

# Preprocess the image and prepare it for the model
inputs = transforms(img, return_tensors='pt')

# Get the model output (logits)
output = model(**inputs)

# Step 4: Convert logits to probabilities
proba = output.logits.softmax(1)

# Convert the probability tensor to a dictionary for easier interpretation
age_ranges = ['0-2', '3-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', 'more than 70']
predictions = {age_range: f"{proba[0, idx].item():.5f}" for idx, age_range in enumerate(age_ranges)}

# Display the predictions
print(predictions)

# Step 5: Find the percentage of the age range "3-9"
age_3_9_percentage = predictions['3-9']
print(f"The percentage of the person being between the age of 3-9 is: {age_3_9_percentage}")


answer = age_3_9_percentage
# ---- End of your code ----
````

In this second task, you will create a user-friendly interface using Gradio for your image classification pipeline that you created in Task 1. The difference with task 1 is, that in this task, you use **image files as input**, process them through the Hugging Face model, and display predictions output. The output displayed is **only the results with the highest `score`**.

Here are the key steps that you might be able to follow:

````python
# @title #### 02. Image Classification using Hugging Face's Model and Gradio
!pip install gradio
# Put your code here:
import gradio as gr
# BACKEND
# Load the pre-trained model and feature extractor
model = ViTForImageClassification.from_pretrained('nateraw/vit-age-classifier')
transforms = ViTFeatureExtractor.from_pretrained('nateraw/vit-age-classifier')

# Function to classify image and return the age range with the highest score
def classify_image(img):
    # Preprocess the image
    inputs = transforms(img, return_tensors='pt')

    # Get the model output (logits)
    output = model(**inputs)

    # Convert logits to probabilities
    proba = output.logits.softmax(1)

    # Define age ranges corresponding to model output
    age_ranges = ['0-2', '3-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', 'more than 70']

    # Find the index with the highest probability
    max_idx = proba.argmax().item()

    # Get the age range and score with the highest probability
    highest_score_age_range = age_ranges[max_idx]
    highest_score = proba[0, max_idx].item()

    # Format the result to display
    result = f'{{\nscore: {highest_score},\nlabel: "{highest_score_age_range}"\n}}'
    return result

# UI
# Create Gradio Interface
iface = gr.Interface(
    fn=classify_image,  # Function to be executed
    inputs=gr.Image(type="pil"),  # Image input, type is 'pil' (PIL Image object)
    outputs=gr.Textbox(),  # Output will be a text string
    title="Age Range Classifier",  # Title of the Gradio app
    description="Upload an image to predict the age range of the person in the image.",  # Description of the app
)

# Launch the interface
iface.launch()



# ---- End of your code ----
````

### Step 4: Preprocess the Data

You need to define the transformation function for image preprocessing and apply it to the dataset.

````python
from transformers import AutoImageProcessor
from torchvision.transforms import RandomResizedCrop, Compose, Normalize, ToTensor

# Load the image processor, Define the transforms, Define the transformation function and Apply the transformation function

# Put your code here:
checkpoint = "google/vit-base-patch16-224-in21k"
image_processor = AutoImageProcessor.from_pretrained(checkpoint)

from torchvision.transforms import RandomResizedCrop, Compose, Normalize, ToTensor

normalize = Normalize(mean=image_processor.image_mean, std=image_processor.image_std)
size = (
    image_processor.size["shortest_edge"]
    if "shortest_edge" in image_processor.size
    else (image_processor.size["height"], image_processor.size["width"])
)


_transforms = Compose([RandomResizedCrop(size), ToTensor(), normalize])

def transforms(examples):
    examples["pixel_values"] = [_transforms(img.convert("RGB")) for img in examples["image"]]
    del examples["image"]
    return examples

emotion = emotion.with_transform(transforms)
# ---- End of your code ----
````

### Step 5: Model Setup

In this step, define the model architecture with the pre-trained ViT model and load it.

````python
from transformers import AutoModelForImageClassification

# Define the model

# Put your code here:
model = AutoModelForImageClassification.from_pretrained(
    checkpoint,
    num_labels=len(labels),
    id2label=id2label,
    label2id=label2id,
# ---- End of your code ----
````

### Step 6: Training Setup

Define the training arguments and instantiate the trainer.

````python
import evaluate

# Load accuracy metric

# Put your code here:
accuracy = evaluate.load("accuracy")
# ---- End of your code ----
````

````python
# Define compute metric function

# Put your code here:
import numpy as np

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return accuracy.compute(predictions=predictions, references=labels)
# ---- End of your code ----
````

````python
from transformers import TrainingArguments, Trainer

# Define training arguments
training_args = TrainingArguments(
    output_dir="model/image_classification",
    remove_unused_columns=False,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=1e-5,                 # Lower learning rate
    per_device_train_batch_size=64,      # Reduced batch size
    gradient_accumulation_steps=2,      # Adjusted gradient accumulation
    per_device_eval_batch_size=64,
    num_train_epochs=50,                # Increased epochs
    warmup_ratio=0.2,                   # Increased warmup ratio
    logging_steps=3,
    load_best_model_at_end=True,        # Early stopping
    metric_for_best_model="accuracy",
    push_to_hub=True,
)

# Update your code here:
# ---- End of your code ----
````

````python
from transformers import DefaultDataCollator

# Instantiate the trainer

# Update your code here:
data_collator = DefaultDataCollator()
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=emotion["train"],
    eval_dataset=emotion["test"],
    tokenizer=image_processor,
    compute_metrics=compute_metrics,
)

# ---- End of your code ----
````

````python
import gradio as gr
from transformers import pipeline

# Load the two models
model_1 = pipeline("image-classification", model="fathurim/image_classification")
model_2 = pipeline("image-classification", model="nateraw/vit-age-classifier")

# Define prediction functions for both models
def predict(image):
    # Predict with the first model
    result_model_1 = model_1(image)[0]
    label_model_1 = result_model_1['label']
    score_model_1 = result_model_1['score']
    result_model_1_formatted = f'{{\nscore: {score_model_1},\nlabel: "{label_model_1}"\n}}'

    # Predict with the second model
    result_model_2 = model_2(image)[0]
    label_model_2 = result_model_2['label']
    score_model_2 = result_model_2['score']
    result_model_2_formatted = f'{{\nscore: {score_model_2},\nlabel: "{label_model_2}"\n}}'

    return result_model_1_formatted, result_model_2_formatted

# Create Gradio interface
interface = gr.Interface(
    fn=predict,  # Function to call for prediction
    inputs=gr.Image(type="pil"),  # Input is a PIL image
    outputs=[gr.Textbox(label="Model 1 Output"), gr.Textbox(label="Model 2 Output")],  # Two text outputs
    title="Dual Model Image Classification"
)

# Launch the interface
interface.launch()
````
