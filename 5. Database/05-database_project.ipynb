```python
# 4. Find the top-selling products (by total sales) for each country, limited to top 5
# Example: sql_top_selling_products_by_country_by_sales = {
#    'COUNTRY 1': [('PRODUCT 1', 200000.00), ('PRODUCT 2', 100000.00), ('PRODUCT 3', 95348.00), ('PRODUCT 4', 90000.43), ('PRODUCT 5', 80000.00)],
#    'COUNTRY 2': [('PRODUCT 1', 9000.00), ('PRODUCT 2', 7345.00), ('PRODUCT 3', 6934.23), ('PRODUCT 4', 6800.23), ('PRODUCT 5', 4000.00)],
#    ... to simplify, country 3 and 4 are not shown
#    'COUNTRY 5': [('PRODUCT 1', 7442.84), ('PRODUCT 2', 5175.00), ('PRODUCT 3', 3098.0), ('PRODUCT 4', 3014.34), ('PRODUCT 5', 2857.83)],

```

```python
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect('project_5.db')
cur = conn.cursor()

# Query to find the top-selling products for each country, limited to top 5 products per country
cur.execute('''
    WITH CountrySales AS (
        SELECT Country, SUM(Quantity * UnitPrice) AS TotalRevenueByCountry
        FROM dataset_p5
        GROUP BY Country
        ORDER BY TotalRevenueByCountry DESC
        LIMIT 5
    ),
    ProductSales AS (
        SELECT Country, Description, SUM(Quantity * UnitPrice) AS TotalRevenueByProduct
        FROM dataset_p5
        GROUP BY Country, Description
    )
    SELECT cs.Country, ps.Description, ps.TotalRevenueByProduct
    FROM ProductSales ps
    JOIN CountrySales cs ON ps.Country = cs.Country
    ORDER BY cs.TotalRevenueByCountry DESC, ps.TotalRevenueByProduct DESC;
''')

# Fetch all results
results = cur.fetchall()

# Process results into the desired format
sql_top_selling_products_by_country_by_sales = {}
for country, product, revenue in results:
    if country not in sql_top_selling_products_by_country_by_sales:
        sql_top_selling_products_by_country_by_sales[country] = []
    sql_top_selling_products_by_country_by_sales[country].append((product, revenue))

# Limit to top 5 products for each country
for country in sql_top_selling_products_by_country_by_sales:
    sql_top_selling_products_by_country_by_sales[country] = sql_top_selling_products_by_country_by_sales[country][:5]



# Close the connection
conn.close()
```

```python
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect('project_5.db')
cur = conn.cursor()

# Query to find the top-selling products for each country, limited to top 5 products per country
cur.execute('''
    WITH CountrySales AS (
        SELECT Country, SUM(Quantity) AS TotalQuantityByCountry
        FROM dataset_p5
        GROUP BY Country
        ORDER BY TotalQuantityByCountry DESC
        LIMIT 5
    ),
    ProductSales AS (
        SELECT Country, Description, SUM(Quantity) AS TotalQuantityByProduct
        FROM dataset_p5
        GROUP BY Country, Description
    )
    SELECT cs.Country, ps.Description, ps.TotalQuantityByProduct
    FROM ProductSales ps
    JOIN CountrySales cs ON ps.Country = cs.Country
    ORDER BY cs.TotalQuantityByCountry DESC, ps.TotalQuantityByProduct DESC;
''')

# Fetch all results
results = cur.fetchall()

# Process results into the desired format
sql_top_selling_products_by_country_by_quantity = {}
for country, product, quantity in results:
    if country not in sql_top_selling_products_by_country_by_quantity:
        sql_top_selling_products_by_country_by_quantity[country] = []
    sql_top_selling_products_by_country_by_quantity[country].append((product, quantity))

# Limit to top 5 products for each country
for country in sql_top_selling_products_by_country_by_quantity:
    sql_top_selling_products_by_country_by_quantity[country] = sql_top_selling_products_by_country_by_quantity[country][:5]


# Close the connection
conn.close()
```

### II.2. Set up Elasticsearch Index

The first step you need to do which is **preprocessing, let's keep it simple and simply drop all row that have missing values.**

Next, you need to set up an Elasticsearch index. An Elasticsearch index is like a database in a traditional relational database. An index lets you store, search, and analyze big volumes of data quickly and in near real time.

You'll also need to setup the mappings for the data types, let's do that here.

```python
# Preprocessing here
dataset = pd.read_csv('data.csv', encoding='ISO-8859-1')
new_dataset = dataset.dropna()
# Load your dataset and create an Elasticsearch index
es = Elasticsearch([{'host': 'localhost', 'port': 9200, 'scheme': 'http'}])
es.indices.create(index='dataset_p5_es', ignore=400)
# Hints: use the BulkAPI for faster creation of the index
#import data to elastic
def dataframe_to_dict(new_dataset):
    records = new_dataset.to_dict(orient='records')
    return records

data_dict = dataframe_to_dict(new_dataset)

def generate_actions(data):
    for record in data:
        yield {
            "_index": "dataset_p5_es",
            "_source": record
        }

from elasticsearch.helpers import bulk
#cek
success, failed = bulk(es, generate_actions(data_dict))
print(f"Berhasil mengunggah {success} dokumen, gagal {failed} dokumen")

```

### II.4. Running Elasticsearch queries

Once your e-commerce data is cleaned and ready, let's analyse the data to gain insights that will drive the expansion strategy, this time using Elasticsearch and we'll focus on one specific country, which is Germany.

1. **Find all transactions for a specific country**

   Your company operates globally, but you want to understand better how different countries contribute to your sales. By finding all transactions that happened in a specific country, you can get a clearer picture of your company's reach and performance in that location. In our case, we'll focus on the country Germany.


2. **Find all unique products in Germany**

   Your first task is to understand the breadth of your company's product portfolio in Germany. By determining the total number of unique products you sell, you can assess the scale and diversity of your operations.


3. **Find the top 5 most purchased products in Germany**

   Next up, you'll identify the best-selling products in your portfolio in Germany. Knowing which items are the most purchased can help guide decisions about product focus and marketing efforts. We'll limit the result to 5 most purchased products.


By executing these tasks, you will enable your team to provide data-driven insights which will directly contribute to the strategic decisions about your business expansion. As part of the data engineering team, you're not only maintaining the technical infrastructure - you're shaping the future of the company.


```python
# Write your Elasticsearch Query here

#use the following variable name to store the result from your Elasticsearch Query
es_total_unique_products = 0
es_transactions_in_germany = 0
es_top_products = {} #dictionary where key is product name, value is the quantity sold, see example below

# 1. Find all transactions for a specific country (Germany)
# Example: es_transactions_in_germany = 7584
query_germany = {
    "query": {
        "match":{
            "Country": "Germany"
        }
    }
}

response_germany = es.count(index=index_name, body=query_germany)
es_transactions_in_germany = response_germany['count']

print(f"Jumlah transaksi di Jerman: {es_transactions_in_germany}")
```

```python
# 2. Find all unique products in Germany
# Hints: Use set() to remove duplicate values
# Example: es_total_unique_products = 45345

query_unique_products = {
    "query": {
        "match": {
            "Country": "Germany"  # Field dan nilai yang dicari
        }
    },
    "aggs": {
        "unique_products": {
            "terms": {
                "field": "Description.keyword",
                "size": 100000
            }
        }
    },
    "size": 0  # Jangan ambil dokumen, hanya hasil agregasi
}

response_unique_products = es.search(index=index_name, body=query_unique_products)


buckets = response_unique_products['aggregations']['unique_products']['buckets']
unique_products = {bucket['key'] for bucket in buckets}  # Menggunakan set untuk menghapus duplikat

# Menghitung jumlah produk unik
es_total_unique_products = len(unique_products)

print(f"Jumlah produk unik di Jerman: {es_total_unique_products}")

```

```python
# 3. Find the top 5 most purchased products in Germany
# Example: es_top_products = {'PRODUCT 1': 498, 'PRODUCT 2': 452, 'PRODUCT 3': 342, 'PRODUCT 4': 231, 'PRODUCT 5': 123}
query_top_products = {
    "query": {
        "match": {
            "Country": "Germany"
        }
    },
    "aggs": {
        "top_products": {
            "terms": {
                "field": "Description.keyword",
                "size": 5,
                "order": {
                    "total_quantity": "desc"
                }
            },
            "aggs": {
                "total_quantity": {
                    "value_count": {
                        "field": "Description.keyword"
                    }
                }
            }
        }
    },
    "size": 0
}

# Menjalankan query
response_top_products = es.search(index=index_name, body=query_top_products)

# Mengambil hasil agregasi
buckets = response_top_products['aggregations']['top_products']['buckets']
es_top_products = {bucket['key']: bucket['total_quantity']['value'] for bucket in buckets}

print(f"Top 5 produk yang paling banyak dibeli di Jerman: {es_top_products}")

```
