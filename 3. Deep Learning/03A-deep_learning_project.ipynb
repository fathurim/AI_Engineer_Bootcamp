```python
# You may add any code here to derive your variables
# Please change this

# Mendefinisikan activation function ReLU
def ReLU(x):
    return np.maximum(0, x)

# Mendefinisikan input
input_1 = 4
input_2 = -3

# Mendefinisikan weights dan bias
## Weight dan bias untuk neuron z1
weight_z1 = np.array([4,2])
bias_z1 = -2

##Weight dan bias untuk neuron z2
weight_z2 = np.array([0,-1])
bias_z2 = -6

##Weight dan bias untuk neuron z3
weight_z3 = np.array([2,3])
bias_z3 = 4

## Combine inputs
inputs = np.array([input_1, input_2])

## Compute z1
z1_linear = np.dot(weight_z1, inputs) + bias_z1
z1_output = ReLU(z1_linear)

## Compute z2
z2_linear = np.dot(weight_z2, inputs) + bias_z2
z2_output = ReLU(z2_linear)

## Combine z1 and z2
hidden_outputs = np.array([z1_output, z2_output])

## Compute z3
z3_linear = np.dot(weight_z3, hidden_outputs) + bias_z3
z3_output = ReLU(z3_linear)

output = z3_output

print(f"The output is {output}")

# assignment_id = "09-deep-learning-project"
# question_id = "q2_simple_neural_network"
# submit(student_id, name, assignment_id, str(output), question_id, drive_link)
```

All layers use the ReLU activation function.

What is the output given the following inputs?

$$
X = \begin{bmatrix}

4 \\
\end{bmatrix}



```python
# You may add any code here to derive your variables
import numpy as np

# Define the ReLU activation function
def relu(x):
    return np.maximum(0, x)

# Input vector
X = np.array([1, 2, 4])

# First layer weights and biases
W1 = np.array([
    [0.23, 0.67, 0.12],
    [-0.89, -0.45, 0.78],
    [0.34, 0.56, -0.90],
    [-0.12, 0.34, 0.56],
    [0.78, -0.90, 0.23]
])

b1 = np.array([0.23, -0.89, 0.34, 0.12, -0.78])

# Second layer weights and biases
W2 = np.array([
    [0.23, 0.67, 0.12, 0.45, 0.89],
    [0.12, 0.34, 0.56, 0.78, 0.90]
])

b2 = np.array([1.96, -1.08])

# Third layer weights and biases
W3 = np.array([1.08, -0.16])
b3 = np.array([-2.8])

# Forward pass through the first layer
Z1 = np.dot(W1, X) + b1
A1 = relu(Z1)

# Forward pass through the second layer
Z2 = np.dot(W2, A1) + b2
A2 = relu(Z2)

# Forward pass through the third layer
Z3 = np.dot(W3, A2) + b3
A3 = relu(Z3)


# Please change this
output = A3

print(f"The output is {output}")

# assignment_id = "09-deep-learning-project"
# question_id = "q3_complex_neural_network"
# submit(student_id, name, assignment_id, str(output), question_id, drive_link)
```

## Task 4

Given [the following sheet (sheet name = Project A)](https://docs.google.com/spreadsheets/d/15JWbRFB4k5CNcfD-2hduHHssXIc1SGhNVSpu_30wVAw/edit#gid=180755192)

Make a backpropagation algorithm to train the network. You can refer to AND and NOT sheet. Use sigmoid activation function for all

Hint: it needs 1 layer with 1 neuron

What is the cost at 10th iteration (the iteration start from 1)?

```python
# Please clone the Google Sheet and do your calculation there

# Please change this
import numpy as np

# Data Input dengan tipe float
X = np.array([[0, 0],
              [1, 0],
              [0, 1],
              [1, 1]], dtype=float)

# Bobot dan bias dengan tipe float
W = np.array([[2.0], [1.0]])  # w0 = 2.0, w1 = 1.0
b = 1.0  # Bias

# Target Output dengan tipe float
Y = np.array([[0],
              [1],
              [1],
              [1]], dtype=float)

# Learning rate
learning_rate = 10

# Fungsi Aktivasi Sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Turunan dari Sigmoid
def sigmoid_derivative(a):
    return a * (1 - a)

# Iterasi untuk training
num_iterations = 10

for i in range(num_iterations):
    # Forward Propagation
    Z = np.dot(X, W) + b
    A = sigmoid(Z)

    # Compute cost 
    cost = (Y - A) ** 2
    total_cost = np.sum(cost)
    # Backward Propagation
    # Gradien dari Cost terhadap Output (a)
    dC_dA = 2 * (A - Y)

    # Gradien dari Output (a) terhadap Z
    dA_dZ = sigmoid_derivative(A)

    # Gradien dari Z terhadap W dan Bias
    dZ_dW = X
    dZ_dB = 1.0

    # Gradien dari Cost terhadap Bobot (w0 dan w1)
    dC_dW = np.dot(dZ_dW.T, dC_dA * dA_dZ) / X.shape[0]
    dC_dB = np.sum(dC_dA * dA_dZ * dZ_dB) / X.shape[0]

    # Update bobot (w0, w1) dan bias (b)
    W[0] -= learning_rate * dC_dW[0]  # Update w0
    W[1] -= learning_rate * dC_dW[1]  # Update w1
    b -= learning_rate * dC_dB         # Update bias


cost = round(total_cost, 2)  # Rounding total_cost


# Don't forget to fill in the link to your Google Sheet
link_to_gsheet = "https://docs.google.com/spreadsheets/d/1T70bSdCg27RQR2xgTTDQnUphpbtRiYA2hMll0cYSH-Q/edit?usp=sharing"

print(f"The cost is {cost}")

assignment_id = "09-deep-learning-project"

question_id = "q4_cost_function_cost"
submit(student_id, name, assignment_id, str(cost), question_id)

question_id = "q4_cost_function_gsheet"
submit(student_id, name, assignment_id, str(link_to_gsheet), question_id, drive_link)
```
